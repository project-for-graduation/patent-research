### 推薦

```python
import pandas as pd
import numpy as np
import json

# 讀取word2vec_data
df = pd.read_csv("word2vec_data.csv")
df.head(3)
```

```python
# word2vec轉np模式
df['word2vec'] = df['word2vec'].map(lambda x : np.array(json.loads(x)))
df.head(3)
```

```python
# 查詢
article_id = 'M620327'
df.loc[df["lication_nums"] == article_id]
```

```python

article_embedding = df.loc[df["lication_nums"] == article_id, "word2vec"].iloc[0]
article_embedding
```

```python
from scipy.spatial import distance
df['sim_value'] = df['word2vec'].map(lambda x : 1 - distance.cosine(article_embedding, x))
df[["lication_nums", "titles", "application_dates", "release_dates", "sim_value" ]].head(3)
```

```python
# 推薦前10
df.sort_values(by = 'sim_value', ascending = False).head(10)
```
